```python
import numpy as np
a = np.random.random((10,1))
b = np.sum(a)
c = np.divide(b,10)
print(c)
```

```python
a = np.random.random((20,1))
b = np.std(a, dtype=np.float64)
print(b)
c = np.var(a, dtype=np.float64)
print(c)
```

```python
a = np.random.random((10,20))
b = np.random.random((20,25))
c = np.matmul(a,b).reshape(10,25)
print(c)
sum = np.sum(c)
print(sum)
```

```python
a = np.random.random((10,1))
b = np.negative(a)
c = np.exp(b)
d = np.add(c,1)
res = np.reciprocal(d)
print(res)
```
